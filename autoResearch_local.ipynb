```python
##install modules and import definitions

!pip install selenium
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
# !pip install matplotlib
!pip install pandas
!pip install openpyxl
# !pip install pyperclip
# !pip install pywinauto
!pip install xlsxwriter
import tqdm
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import numpy as np
import matplotlib.pyplot as plt
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl
import time
# import pyperclip
import os
from datetime import datetime
import xlsxwriter
# import pywinauto

def login(Id, password,driver):
    #login at mobile emulater'https://www.instagram.com/'
    #input : Id and password of user, and driver which is webdriver object
    #output : None
    driver.implicitly_wait(4)
    driver.find_element(By.XPATH, '//*[@id="id"]').send_keys(Id)
    driver.find_element(By.XPATH, '//*[@id="pw"]').send_keys(password) 
                                                    
    click_to_login = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="col_pw"]/div')))
    click_to_login.click()
    
    # 중앙대 요금 미납 관련
    try :
        driver.find_element(By.XPATH, '//*[@id="column_5"]/div/div[10]').click()
    except :
        pass
    
def get_back(driver):
    driver.find_element(By.XPATH, '//*[@id="top_btns"]/p[1]/a/img').click()

def set_period(daystart, dayend, driver, change_month=None): ##Added on 2.24
#    need to implement before getting table both in get_by_
#    input :
#        no change if value is -1 both for daystart and dayend
#        daystart(num) : start day to search, 2 if you want to get date 2 on this month
#        dayend(num) : end day to search
#        driver : webdriver

    def click_date(day_to_click,driver) :
        #  input : num (day)
        #  output : None, select day
        #  driver : webdriver
        time.sleep(1)
        startcalendar = driver.find_elements_by_class_name('ui-state-default')
        time.sleep(3)
        for day in startcalendar :
          if int(day.text) == day_to_click :
              day.click()
              break

    if daystart != 0 :

        datepicker_start = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="datepicker2"]')))
        datepicker_start.click()
        if change_month is not None :
          previous_arrow = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="ui-datepicker-div"]/div/a[1]')))
          previous_arrow.click()
          time.sleep(1)
        click_date(daystart, driver)
    if dayend != 0 :
        
        datepicker_end = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="datepicker3"]')))
        datepicker_end.click()
        driver.implicitly_wait(6)
        if change_month is not None :
          previous_arrow = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="ui-datepicker-div"]/div/a[1]')))
          previous_arrow.click()
          time.sleep(1)
        click_date(dayend, driver)
    submit = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="sub1_btns"]/p[1]/a/img')))
    submit.click()
    driver.implicitly_wait(4)
    
def click_get_by_goods(driver): ##Added on 2.24
    
    click_to_ = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="tb_mainicon"]/tbody/tr[2]/td[3]/a/img')))
    click_to_.click()
    
def get_by_goods(driver): ##Updated on 22.2.24
    driver.implicitly_wait(4)
    table = driver.find_element_by_class_name('sub_tb2_2')
    tbody = table.find_element_by_tag_name("tbody")
    rows = tbody.find_elements_by_tag_name("tr")

    row_index=[0]*len(rows)
    col_index = ['분류명', '상품명','가격','건수','합계']
    
    table_array = np.array([[None]*len(col_index) for _ in range(len(row_index))])
    
    for index, value in enumerate(rows):
        body=value.find_elements_by_tag_name("td")
        table_array[index,:] = [element.text for element in body]
    
    return table_array

def click_get_by_time(driver): ##Added on 22.2.24
    
    click_to_ = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="tb_mainicon"]/tbody/tr[1]/td[3]/a/img')))
    click_to_.click()
    
    
def get_by_time(driver): ##Updated on 22.2.24
    
    driver.implicitly_wait(4)
    
    table = driver.find_element_by_class_name('sub_tb2_2')
    tbody = table.find_element_by_tag_name("tbody")
    rows = tbody.find_elements_by_tag_name("tr")

    row_index=[0]*len(rows)
    col_index=['판매시간', '매출합계', '건수', '객수']
    
    table_array = np.array([[None]*len(col_index) for _ in range(len(row_index))])
   
    for index, value in enumerate(rows):
        body=value.find_elements_by_tag_name("td")
        table_array[index,:] = [element.text for element in body]
    return table_array

def fine_data(info_by_goods,name):
    
    # input : np.array of df by goods where one shop
    # output : preprocess of data which is useless( option, etc)
    #         and add rows of same index, and drop all duplicated rows by index
    #         and output fined, sorted dataframe
    #updated(22.02.22) - daily sell divided into delievery or not
    list_to_del=[]
    delievery_company_list = ['쿠팡이츠 매출등록_','쿠팡이츠 매출등록','요기요 매출등록_','요기요 매출등록' ,'배달의민족 매출등록_',
                              '배달의민족 매출등록','헝그리판다 매출등록_','헝그리판다 매출등록']

    for index in range(len(info_by_goods)) :
        if info_by_goods[index][-1]== '0':
            list_to_del.append(index)
        elif info_by_goods[index][0] == '' :
            list_to_del.append(index)
        elif info_by_goods[index][0] == '소계' :
            list_to_del.append(index)

    fined = np.delete(info_by_goods, list_to_del, axis=0)

    temp=pd.DataFrame(fined[:-1,2:],index=fined[:-1,1],columns = ['가격','수량','소계'])

    for col_idx in range(len(temp.columns)):
        temp[temp.columns[col_idx]] = temp[temp.columns[col_idx]].str.replace(',', '').astype('int64')

    for index in temp.index :
        if temp.index.value_counts().loc[index]>1:
            temp.loc['{}_'.format(index)] = temp.loc[index].sum(axis=0)

    fined_df = temp.loc[~temp.index.duplicated(keep=False)]
    delievery_sum = sum([fined_df.loc[i, '소계'] for i in delievery_company_list if i in fined_df.index])
    daily_sum = fined_df['소계'].sum()
    fined_df.loc['배달/현장/합계'] = [delievery_sum,daily_sum-delievery_sum,daily_sum]
    sorted_df = fined_df.sort_values(by='수량', ascending=False)
    sorted_df.index.name = name
    sorted_df.reset_index(level=0,inplace=True)
    
    return sorted_df

def make_filename (daystart, dayend) : #Added on 22.02.24
  ##input : (num, num)
  ##output : filename, if input is 0, it automatcally gives value as today's date
  ##error : if daystart is bigger than dayend, it triggers error
  today = datetime.today().strftime("%Y%m%d%H%M%S")[:8]
  nowdate = today[-2:]
  daystart = daystart if daystart else nowdate
  dayend = dayend if dayend else nowdate
  if int(daystart) <10 :
    daystart = '0' +str(daystart)
  if int(dayend) <10 :
    dayend = '0' + str(dayend)
  if int(daystart) > int(dayend) :
    raise Exception("daystart can never bigger than dayend")
  if int(daystart) > int(nowdate) or int(dayend) > int(nowdate) :
    raise Exception("Range can't bigger than today's date which is day {}".format(nowdate))
  filename = today[:-2] + str(daystart) + '-' + str(dayend)

  return filename

def logout():
    logout = driver.find_element_by_xpath('//*[@id="top_btns"]/p[3]/a')
    logout.click()
    time.sleep(0.5)
    
    for _ in range(2):
        driver.switch_to.alert.accept()
def chromeSetting():
    ## setting chrome options for colab env
    ## return : Setted driver for colab
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    return driver
```

```python
def input_by_pyperclip(tag_id, tag_pwd, Id, pwd):
    
    ## input : (webdrivertag_Id and pwd, Id, pwd)
    ## output : input Id and pwd with pyperclip
     # id 입력
#     tag_id.clear()
    
    tag_id.click()
    pyperclip.copy(Id)
    tag_id.send_keys(Keys.CONTROL, 'v')
    time.sleep(1)

    # pw 입력
    tag_pwd.click()
    pyperclip.copy(pwd)
    tag_pwd.send_keys(Keys.CONTROL, 'v')
    time.sleep(1)
    
def check_review(driver):
    
    ans = {'배민':0, '요기요':0}
    mention = '맛있게 드시고 리뷰 남겨주셔서 감사드립니다! \n\n이번에도 다음에도 맛있는 세계여행 제공해드리겠습니다! 무엇보다 건강조심하시고 다음에 또 뵈요'
    driver = webdriver.Chrome()
    driver.get('https://ceo.baemin.com/web/login?returnUrl=https%3A%2F%2Fceo.baemin.com%2Fself-service&__ts=1607756604872')
    

    tag_id = driver.find_element_by_id('id')
    tag_pw = driver.find_element_by_id('pw')

    input_by_pyperclip(tag_id, tag_pw, 'bebridgeewha','sangin6262')
    
    driver.find_element_by_id('btnLogin').click()
    time.sleep(1)
    
    driver.find_element_by_xpath('//*[@id="root"]/div/div[1]/div/div[2]/div/span/a').click()
    time.sleep(2.5)
    driver.find_element_by_xpath('//*[@id="root"]/div/div[1]/div[1]/button').click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="root"]/div/div[1]/div[2]/div/nav/ul/li[2]/ul/li[1]/a').click()
    time.sleep(1)
    state_review = driver.find_element_by_xpath('//*[@id="review-page"]/div[3]/label[2]/p')
    
    ans['배민']= state_review.text
    
    no_reply = int(ans['배민'][4])
    inputs_ = driver.find_elements_by_xpath('//*[@id="review-page"]/div[4]/div/div/div[2]/div/button')

    for i in range(no_reply):
        inputs_[i].click()
    
    time.sleep(1)
    
#     text = driver.find_elements_by_xpath('//*[@id="review-page"]/div[4]/div/div/div[2]/div/div/div[1]/textarea')
#     for i in range(no_reply):
#         text[i].send_keys(mention)
    
    yogiyo = 'https://owner.yogiyo.co.kr/owner/'
    driver.execute_script('window.open("{}");'.format(yogiyo))
    driver.switch_to_window(driver.window_handles[1])
    time.sleep(0.5)

    tag_id = driver.find_element_by_id('username')
    tag_pw = driver.find_element_by_id('password')
    input_by_pyperclip(tag_id,tag_pw, 'bebridgeewha','sangin6262~')
    time.sleep(1)
    driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div[2]/div/form/div[5]/button').click()
    time.sleep(1)#login
    driver.find_element_by_xpath('//*[@id="modal-pwd-notice"]/div[3]/a[1]').click()
    time.sleep(1)#pass change pwd
    driver.find_element_by_xpath('//*[@id="nav"]/div/ul[1]/li[1]/a').click()
    time.sleep(1)#업소관리
    driver.find_element_by_xpath('/html/body/div[2]/ul/li[13]/a').click()
    time.sleep(1)#review
    all_review = driver.find_element_by_xpath('//*[@id="main"]/div[2]/div[3]/div[1]/ul/li[1]/strong')
    had_reply = driver.find_element_by_xpath('//*[@id="main"]/div[2]/div[3]/div[1]/ul/li[2]/strong')

    ans['요기요'] = '미답변({})'.format(int(all_review.text)-int(had_reply.text))
    
    time.sleep(1)
    no_reply = int(ans['요기요'][4])
    text = driver.find_elements_by_xpath('//*[@id="main"]/div[2]/div[3]/ul/li/div[2]/div/div[1]/textarea')
    for i in range(no_reply):
        text[i].send_keys(mention)
        
    print(ans)
    return driver


def login(Id, password,driver):
    #login at mobile emulater'https://www.instagram.com/'
    #input : Id and password of user, and driver which is webdriver object
    #output : None
    driver.implicitly_wait(4)
    driver.find_element_by_xpath('//*[@id="id"]').send_keys(Id)
    driver.find_element_by_xpath('//*[@id="pw"]').send_keys(password) 
                                                    
    click_to_login = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="col_pw"]/div')))
    click_to_login.click()
    
    # 중앙대 요금 미납 관련
    try :
        driver.find_element_by_xpath('//*[@id="column_5"]/div/div[10]').click()
    except :
        pass
    
def get_back(driver):
    driver.find_element_by_xpath('//*[@id="top_btns"]/p[1]/a/img').click()

def get_by_goods(driver):
    click_to_ = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="tb_mainicon"]/tbody/tr[2]/td[3]/a/img')))
    click_to_.click()
    
    driver.implicitly_wait(4)
    
    table = driver.find_element_by_class_name('sub_tb2_2')
    tbody = table.find_element_by_tag_name("tbody")
    rows = tbody.find_elements_by_tag_name("tr")

    row_index=[0]*len(rows)
    col_index = ['분류명', '상품명','가격','건수','합계']
    
    table_array = np.array([[None]*len(col_index) for _ in range(len(row_index))])
    
    for index, value in enumerate(rows):
        body=value.find_elements_by_tag_name("td")
        table_array[index,:] = [element.text for element in body]
    
    return table_array

def fine_data(info_by_goods,name):
    
    # input : np.array of df by goods where one shop
    # output : preprocess of data which is useless( option, etc)
    #         and add rows of same index, and drop all duplicated rows by index
    #         and output fined, sorted dataframe
    
    list_to_del=[]

    for index in range(len(info_by_goods)) :
        if info_by_goods[index][-1]== '0':
            list_to_del.append(index)
        elif info_by_goods[index][0] == '' :
            list_to_del.append(index)
        elif info_by_goods[index][0] == '소계' :
            list_to_del.append(index)

    fined = np.delete(info_by_goods, list_to_del, axis=0)
    temp=pd.DataFrame(fined[:-1,2:],index=fined[:-1,1],columns = ['가격','수량','소계'])

    for col_idx in range(len(temp.columns)):
        temp[temp.columns[col_idx]] = temp[temp.columns[col_idx]].str.replace(',', '').astype('int64')

    for index in temp.index :
        if temp.index.value_counts().loc[index]>1:
            temp.loc['{}_'.format(index)] = temp.loc[index].sum(axis=0)

    fined_df = temp.loc[~temp.index.duplicated(keep=False)]
    fined_df.loc['상품명'] = ['합계',fined_df['수량'].sum(),fined_df['소계'].sum()]
    sorted_df = fined_df.sort_values(by='수량', ascending=False)
    sorted_df.index.name = name
    sorted_df.reset_index(level=0,inplace=True)
    
    return sorted_df

def get_by_time(driver):
    click_to_ = WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="tb_mainicon"]/tbody/tr[1]/td[3]/a/img')))
    click_to_.click()
    
    driver.implicitly_wait(4)
    
    table = driver.find_element_by_class_name('sub_tb2_2')
    tbody = table.find_element_by_tag_name("tbody")
    rows = tbody.find_elements_by_tag_name("tr")

    row_index=[0]*len(rows)
    col_index=['판매시간', '매출합계', '건수', '객수']
    
    table_array = np.array([[None]*len(col_index) for _ in range(len(row_index))])
   
    for index, value in enumerate(rows):
        body=value.find_elements_by_tag_name("td")
        table_array[index,:] = [element.text for element in body]
    
    return table_array

def logout():
    logout = driver.find_element_by_xpath('//*[@id="top_btns"]/p[3]/a')
    logout.click()
    
    time.sleep(0.5)
    
    for _ in range(2):
        driver.switch_to.alert.accept()
def chromeSetting():
    ## setting chrome options for colab env
    ## return : Setted driver for colab
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    return driver
def send_email(mail_nm, what, *whos):
    
    ##input : name of mail, to who, to uploadwhat
    ##output : send file of 'what' to 'who' with name of 'mail_nm'

    # 로그인 버튼을 찾고 클릭합니다
    driver = chromeSetting()
    driver.get('https://www.naver.com/')
    
    login_btn = driver.find_element_by_class_name('link_login')
    login_btn.click()
    time.sleep(1)

    # id, pw 입력할 곳을 찾습니다.
    tag_id = driver.find_element_by_name('id')
    tag_pw = driver.find_element_by_name('pw')
    
    input_by_pyperclip(tag_id, tag_pw, 'ie1914', '19921214.Z')

    # 로그인 버튼을 클릭합니다
    login_btn = driver.find_element_by_id('log.login')
    login_btn.click()

    driver.implicitly_wait(4)

    driver.find_element_by_xpath('//*[@id="frmNIDLogin"]/fieldset/span[2]').click()
    driver.find_element_by_xpath('//*[@id="NM_FAVORITE"]/div[1]/ul[1]/li[1]').click()
    driver.find_element_by_xpath('//*[@id="nav_snb"]/div[1]/a[1]/strong').click()

    time.sleep(1)
    
    address = driver.find_element_by_xpath('//*[@id="toInput"]')
    
    for who in whos :  
        address.send_keys(who,Keys.ENTER)
    
    driver.find_element_by_xpath('//*[@id="subject"]').send_keys(mail_nm)
    
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="AddButton_html5"]/span').click()
    
    time.sleep(2)
    app = pywinauto.Application().connect(title = '열기')
    app.열기.Edit.set_edit_text(what)
    time.sleep(1)
    app.열기.Button.click()
    time.sleep(2)

    driver.find_element_by_xpath('//*[@id="sendBtn"]').click()
    
def back_up(file_name): 
    
    today = datetime.today().strftime("%Y%m%d%H%M%S")[:8]
    this_file = os.path.join(os.getcwd(),file_name)
    send_email('Backup_{}'.format(today),this_file,'ie1914@naver.com')
```
